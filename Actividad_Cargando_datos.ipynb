# Cargando datos

lo primero que necesitamos hacer para poder visualizar los datos es cargar los datos en nuestro programa. En este notebook vamos a cargar los datos que ya nos hemos descargado en nuestro directorio. 

Si miramos los datos nos damos cuenta de que los datos están guardados en un formato conocido como **.json** Por tanto necesitamos una manera de poder leer este tipo de ficheros. Si buscamos en google nos damos cuenta de que la librería adecuada para hacerlo se llama **json**. Importemos dicha librería.

```python
#Esta es una librería específica para cargar datos en formato json
import json
```

```python
with open('Statsbomb/data/competitions.json') as f:
    competitions = json.load(f)
```

```python
# Veamos que se ha cargado en nuestra variable competitions. Es una estructura que en Python se llama diccionario, 
# tiene unas claves (keys): competition_id, season_id, country_name, .... y unos valores asociados a esas claves:
# 72, 30, 'International'...

competitions
```

```python
# Vamos por tanto a elegir una competición. Por ejemplo la última que tiene una competition_id: 72

#Womens World Cup 2019 has competition ID 72
competition_id=72
```

```python
# Una vez elegida la competición podemos cargar todos los partidos que hay asociadas a esa competición

#with open('Statsbomb/data/matches/72/30.json') as f:



with open('Statsbomb/data/matches/'+str(competition_id)+'/30.json') as f:
    matches = json.load(f)
```

Como vemos podemos buscar la información que necesitamos en nuestra variable de partidos matches. Pero lo bueno de saber programar es que podemos hacer las cosas de manera automática. Por ejemplo pensemos que **queremos saber cuales fueron todos los partidos y sus resultados**. Necesitamos por tanto recorrer todos los datos que hay en matches y sacar la información que queremos.

```python
for partido in matches:
    print(partido)
    
# Como vemos me imprime todo lo que hay pasando uno por uno. 
# Vamos a extraer lo que necesitamos: Nombre de los equipos y sus resultados
```

```python
for match in matches:
    
    home_team_name = match['home_team']['home_team_name']
    away_team_name = match['away_team']['away_team_name']
    home_score = match['home_score']
    away_score = match['away_score']
    
    describe_text = 'El partido entre '+ home_team_name + ' y ' + away_team_name 
    result_test =' acabó ' + str(home_score)+':'+ str(away_score)
    
    print(describe_text + result_test)

# Aquí hemos hecho un par de cosas nuevas. Primero concatenar en el print, simplemente poniendo un + 
# y además como teníamos que concatenar strings y números hemos convertido los números en cadenas de caracteres (strings)
```

Supongamos ahora que estamos interesados en extraer la información de un determinado partido, como por ejemplo el de **España contra Estados Unidos**

```python
home_team_required ="Spain Women's"
away_team_required ="United States Women's"

# Find the ID of the match

for match in matches:
    home_team_name = match['home_team']['home_team_name']
    away_team_name = match['away_team']['away_team_name']
    if (home_team_name == home_team_required) and (away_team_name==away_team_required):
        match_id_required = match['match_id']
print('this match has the match_id = ',match_id_required)
```
